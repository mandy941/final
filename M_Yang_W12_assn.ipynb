### create and read xml file

```python
# load the urllib.request function 
import urllib.request
# load the objectify() function from the lxml library
from lxml import objectify
```

```python
# open the convertcsv.xml web page containing the data set, 
# capturing the header info separately from the path
path, headers = urllib.request.urlretrieve('https://raw.githubusercontent.com/mandy941/final/master/convertcsv.xml')

# use objectify() to parse the web page
parsed = objectify.parse(open(path))

# get a reference to the root node of the convertcsv.xml file
root = parsed.getroot()

# show the root node of the convertcsv.xml file
root
```

```python
# show the children of the root node
root.getchildren()
```

```python
# define an empty list to store the parsed data
data = []
# use root.row to extract each <row> element from convertcsv.xml
for elt in root.row:
     # for each row, create a dict of tag names
    tag_data = {}
    for child in elt.getchildren():
        tag_data[child.tag] = child.pyval
    data.append(tag_data)
    
data
```

```python
# use pandas function to make the file as dataframe
# print the result
xml = pd.DataFrame(data)
xml
```
