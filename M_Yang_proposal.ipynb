# Statistical Analysis And Graphics

###Analysis for data sources one

```python
#remove the unmeaningful coulumn from the data set one,name df0
df2 = df0.drop(columns = ['Group Name/Partner','LocationType', 'Event Type','Category','Classification'])
df2
```

```python
# use groupby function to group the areas of NYC and count the special event types times for each area 
location = df2.groupby('Borough')[['Unit']].count()
location
```

There are five areas have special park events in NYC, including Bronx, Brooklyn, Manhattan, Queens, Staten Island.

```python
# use the bar chart to see the special park events' comparesion among different areas in NYC
location['Unit'].plot(kind = 'bar')
```

The Mahattan has the highest number special park events。

```python
# use the bar chart to see the frequency of each areain NYC
location['Unit'].plot(kind = 'hist')
```

According to the frequency by histigram, we can know that the main range of park special events is from 100 to 250.

```python
#use groupby function to group the organizations and count the special event types times for each organization 
unit = df2.groupby('Unit')[['Borough']].count()
unit1 = unit.T
unit1
```

```python
#get descriptive statistics for the number of special park events in each organizations by describe function
unit.describe()
```

we can know that the number of all organizations performe the average special park events are 210. One of the organization only performe 1 events(min) and the maximum is 995.

```python
# use the bar chart to see the special park events' comparsion among different organizations 
unit['Borough'].plot(kind = 'bar')
```
