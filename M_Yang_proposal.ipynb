# Statistical Analysis And Graphics

```python
# use groupby function to group the areas of NYC and count the special event types times for each area 
location = df0.groupby('Borough')[['Unit']].count()
location
```

There are five areas have special park events in NYC, including Bronx, Brooklyn, Manhattan, Queens, Staten Island.

```python
# use the bar chart to see the special park events' comparesion among different areas in NYC
location['Unit'].plot(kind = 'bar')
```

The Mahattan has the highest number special park events。

```python
# use the bar chart to see the frequency of each areain NYC
location['Unit'].plot(kind = 'hist')
```

According to the frequency by histigram, we can know that the main range of park special events is from 100 to 250.
